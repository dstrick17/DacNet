```python

import matplotlib.pyplot as plt


top_14 = data['Finding Labels'].value_counts().head(20)

# Plotting the top 20 categories
plt.figure(figsize=(10, 6))
top_14.plot(kind='bar', color='skyblue')
plt.title('Top 14 Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
